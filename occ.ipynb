```python
import pandas as pd

occ_data =pd.read_csv('/Users/jetto/py/occ_title_dic.csv').dropna(how='all')
occ_data
```

```python
wob = occ_data['2009occ_title'].dropna(how='any').tolist()
wob = [str.lower(str(s)).replace('and','').replace(',','').split() for s in wob]
wob = [list(set(s)) for s in wob]
```

```python
wob[125]
```

```python
import numpy as np

vec = pd.DataFrame()
k = 0
for l in wob:
    temp = pd.DataFrame(1,index=l,columns=['No'+str(k)])
    try:
        vec = pd.concat([vec,temp],axis=1,join='outer')
    except:
        d1 = vec
        d2 = temp
        break
    k = k + 1
#d1 = pd.DataFrame(1,index=list(d1.index),columns=['d1'])
#d2 = pd.DataFrame(1,index=list(d2.index),columns=['d2'])
vec = vec.fillna(0)
```

```python
wob2 = occ_data['2019occ_title'].dropna(how='any').tolist()
wob2 = [str.lower(str(s)).replace('and','').replace(',','').split() for s in wob2]
wob2 = [list(set(s)) for s in wob2]
```

```python
vec2 = pd.DataFrame()
k = 0
for l in wob2:
    temp = pd.DataFrame(1,index=l,columns=['2019_'+str(k)])
    try:
        vec2 = pd.concat([vec2,temp],axis=1,join='outer')
    except:
        d1 = vec2
        d2 = temp
        break
    k = k + 1
#d1 = pd.DataFrame(1,index=list(d1.index),columns=['d1'])
#d2 = pd.DataFrame(1,index=list(d2.index),columns=['d2'])
vec2 = vec2.fillna(0)
```

```python
temp = pd.concat([vec,vec2],axis=1)
vec = temp.iloc[:,:vec.shape[1]].fillna(0)
vec2 = temp.iloc[:,vec.shape[1]:].fillna(0)
```

```python
for i in range(vec.shape[1]):
    vec.iloc[:,1] = vec.iloc[:,1]/np.linalg.norm(vec.iloc[:,1])
```

```python
for i in range(vec2.shape[1]):
    vec2.iloc[:,1] = vec2.iloc[:,1]/np.linalg.norm(vec2.iloc[:,1])
```

```python
temp = vec.T@vec2
address = []
for i in range(temp.shape[1]):
    s = temp.iloc[:,i]
    idx = list(s[s==s.max()].index)
    address = address + [idx]
```
