```python
import requests
from bs4 import BeautifulSoup

import itertools

url = 'https://www.federalreserve.gov/monetarypolicy/fomcprojtabl20201216.htm'

html = requests.get(url)
soup = BeautifulSoup(html.content, "html.parser")
soup
```

```python
elems = soup.select('tr')
```

```python
data = []
chk = 0
for s in elems:
    if ('Midpoint of target range' in s.text) or (chk==1):
        ss = s.text.replace('\xa0','0').replace('\n',',')
        data = data + [ss[1:-1].split(',')]
        chk = 1
    if 'class="stub" headers="1f92d6f0" id="f00ad4c8"' in str(s):
        break
        
import pandas as pd
df = pd.DataFrame(data[1:],columns=data[0])
df = df.set_index('Midpoint of target range or target level (Percent)',drop=True)
df
```

```python
element = soup.find_all('div', class_='data-table')
len(element)
```

# GDP成長率

```python
elems = element[6].find_all('tr')
elems = [s.text.replace('\xa0','0').replace('\n',',') for s in elems]
#elems
```

```python
s1 = elems[0].split(',')[1:-1]
s2 = [[ss+'sep',ss+'dec']for ss in s1[1:]]
s2 = list(itertools.chain.from_iterable(s2))
s1 = ['GDP'] + s2
#s1
```

```python
out = [s1]
out = out + [s.split(',')[1:-1] for s in elems[2:]]
out = pd.DataFrame(out[1:],columns=out[0])
out = out.set_index('GDP',drop=True)
out
```
